# Estadísticas descriptivas

```python
# Gráficos compartivos
# Primero veamos la distribución de la variable objetivo 'Calories_Burned'
plt.figure(figsize=(8, 5))
sns.histplot(df['Calories_Burned'], bins=30, kde=True, color='skyblue')
plt.title('Distribución de Calories_Burned')
plt.xlabel('Calories_Burned')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Ahora hacemos el gráfico para visualizar la distribución entre las columnas numéricas
columnas_numericas = df.select_dtypes(include=['number']).columns.tolist()

# Creamos los subplots
fig,axes = plt.subplots(nrows=len(columnas_numericas), ncols=1, figsize=(8, 5*len(columnas_numericas)))

# Aplanamos los ejes para iterar fácilmente
axes = axes.flatten()

# Creamos un histograma para cada columna numérica
for i, col in enumerate(columnas_numericas):
    sns.histplot(df[col], bins=30, kde=True, ax=axes[i], color='skyblue')
    axes[i].set_title(f'Distribución de {col}')
    axes[i].set_xlabel(col)
plt.tight_layout()
plt.show()

```

```python
# Ahora hacemos el gráfico para ver la comparación entre el genero y las calorías quemadas
plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='Calories_Burned', data=df)
plt.title('Calorías Quemadas por Género')
```

```python
# Ahora vamos con la matriz de correlación de las variables numéricas
matriz_correlacion = df[columnas_numericas].corr()

# Mostramos la matriz de correlación
plt.figure(figsize=(12, 8))
sns.heatmap(matriz_correlacion, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Matriz de Correlación de Variables Numéricas')
plt.show()
```
