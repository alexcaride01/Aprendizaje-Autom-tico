# Estadísticas Descriptivas Clasificación

```python
# Gráficos Comparativos
plt.figure(figsize=(8, 5))
sns.histplot(df['respuesta'], bins=10, kde=False, color='orange')
plt.title('Distribución de la variable "respuesta"')
plt.xlabel('Respuesta (0 = No, 1 = Sí)')
plt.show()
```

```python
# Seleccionamos las columnas numéricas
columnas_numericas = df.select_dtypes(include=['number']).columns.tolist()
print(f"Columnas numéricas: {columnas_numericas}")

# Creamos los subplots (uno por variable numérica)
fig, axes = plt.subplots(
    nrows=len(columnas_numericas), 
    ncols=1, 
    figsize=(8, 5 * len(columnas_numericas))
)

# Si solo hay una columna numérica, axes no será una lista
if len(columnas_numericas) == 1:
    axes = [axes]

# Iteramos sobre cada columna numérica para graficar
for i, col in enumerate(columnas_numericas):
    sns.histplot(df[col].dropna(), bins=30, kde=True, ax=axes[i], color='skyblue')
    axes[i].set_title(f'Distribución de {col}')
    axes[i].set_xlabel(col)


plt.tight_layout()
plt.show()
```

```python
# Ahora vamos con la matriz de correlación de las variables numéricas
matriz_correlacion = df[columnas_numericas].corr()

# Mostramos la matriz de correlación
plt.figure(figsize=(12, 8))
sns.heatmap(matriz_correlacion, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Matriz de Correlación de Variables Numéricas')
plt.show()
```
